## Square Max-Cut

```python
import dimod
import greedy
J = {(0,1):1, (1,2):1,(2,3):1,(3,0):1}
h = {}
problem = dimod.BinaryQuadraticModel(h, J, 0.0, dimod.SPIN)
solver = greedy.SteepestDescentSolver()
solution = solver.sample(problem, num_reads = 10000) 
solution.aggregate().to_pandas_dataframe().sort_values('num_occurrences',ascending=False).reset_index(drop=True)
```

## 1D heisenberg chain Ferro

```python
import dimod
import greedy
length_1D= 10
J_spin={(i,i+1):-1 for i in range(20)}
J_spin[(length_1D,0)]=-1
N_reads=1000000

J_spin
h = {}
problem = dimod.BinaryQuadraticModel(h, J_spin, 0.0, dimod.SPIN)
solver = greedy.SteepestDescentSolver()
solution = solver.sample(problem, num_reads = N_reads) 
solution_df = solution.aggregate().to_pandas_dataframe().sort_values('num_occurrences',ascending=False).reset_index(drop=True)
print((solution_df["num_occurrences"][0]+solution_df["num_occurrences"][1])/N_reads)
solution_df
```
