```python
from qiskit.opflow import Z
from qiskit.opflow import I
from qiskit.algorithms import QAOA
from qiskit.utils import algorithm_globals, QuantumInstance
from qiskit import Aer
from qiskit.algorithms.optimizers import COBYLA
import pandas as pd 
def return_chain(N_chain,value = 1):
    Total = 0
    for j in range(N_chain):
        temp=1
        for k in range(N_chain):
            if k==j or k==(j+1)%N_chain:
                temp=temp^Z
            else:
                temp=temp^I
        Total += (value*temp)
    return Total
```

```python
H1 = return_chain(15,1)
reps_value = 500
constrains_list = [i+1 for i in range(reps_value*2) ]
qaoa = QAOA(reps = reps_value)
circuit = qaoa.construct_circuit(constrains_list,H1)
quantum_instance = QuantumInstance(Aer.get_backend("aer_simulator"), shots = 1024*2)
qaoa = QAOA(optimizer = COBYLA(), quantum_instance=quantum_instance)
result = qaoa.compute_minimum_eigenvalue(H1)
sorted_result = sorted(result.eigenstate.items(), key=lambda x:x[1],reverse=True)
print(sorted_result[:3])
print(result)
```
